```python
from abacusai import PredictionClient

def get_chat_response(deployment_token, deployment_id, user_message):
    client = PredictionClient()
    response = client.get_chat_response(
        deployment_token=deployment_token,
        deployment_id=deployment_id,
        messages=[{"is_user": True, "text": user_message}],
        llm_name=None,
        num_completion_tokens=None,
        system_message=None,
        temperature=0.0,
        filter_key_values=None,
        search_score_cutoff=None,
        chat_config=None
    )
    # Assuming the response is in the format of a list of messages
    # Extract the text from the first message (assuming it's the AI's response)
    print("Response:", response)
    response_text = response.get('messages', [])[0].get('text', 'No response received.')

    return response

# Example usage
deployment_token = 'ed67ab9c0a4b4d0f9d4c246e84d7be46'
deployment_id = '7eed26856'
user_message = "summarise week 5"
response = get_chat_response(deployment_token, deployment_id, user_message)
print(response)
```
