```python
import os 
import sys 

import psycopg2
database = sys.argv[1]	

user = os.environ.get('PGUSER') 

password = os.environ.get('PGPASSWORD') 

host = os.environ.get('PGHOST')

port = os.environ.get('PGPORT')



import psycopg2

def shift_cipher(char):
  if char.isdigit():
      return str((int(char) + 7) % 10)
  elif char.isalpha():
      if char.isupper():
          return chr((ord(char) - ord('A') + 7) % 26 + ord('A'))
      else:
          return chr((ord(char) - ord('a') + 7) % 26 + ord('a'))
  return char

def encode_id(team_id):
  return ''.join(shift_cipher(char) for char in team_id)

def connectDb(database="FLIS", username="myuser", pwd="MYPASS", 
            address="127.0.0.1", portnum="5432"):
  conn = None
  try:
      # connect to the PostgreSQL database
      conn = psycopg2.connect(database=database, user=username, 
                              password=pwd, host=address, port=portnum)
      curr = conn.cursor()
      
      curr.execute("SELECT team_id FROM teams WHERE jersey_home_color <> jersey_away_color")
      data = curr.fetchall()
      
      # Extract team_ids and sort them
      team_ids = sorted([team_id[0] for team_id in data])
      
      # Encode each team_id
      encoded_ids = [encode_id(team_id) for team_id in team_ids]
      
      # Print each encoded team_id
      for encoded_id in encoded_ids:
          print(encoded_id)
      
  except (Exception, psycopg2.DatabaseError) as error:
      print(error)
  finally:
      if conn is not None:
          conn.close()

# Call the function
connectDb()
```

```python
import os
import sys
import math
import psycopg2

# Database connection parameters
database = sys.argv[1]
user = os.environ.get('PGUSER')
password = os.environ.get('PGPASSWORD')
host = os.environ.get('PGHOST')
port = os.environ.get('PGPORT')

def calculate(result):
    
    # If result is None, set sum_scores to 0
    sum_scores = result if result is not None else 0
    
    # Compute X
    X = sum_scores * 10
    
    # Convert X from radians to degrees
    X_deg = X * (math.pi / 180)
    
    # Calculate the cosine of X_deg
    cos_X_deg = round(math.cos(X_deg), 2)
    return cos_X_deg

def connect_db(database, username, pwd, address, portnum):
  conn = None
  try:
      # Connect to the PostgreSQL database
      conn = psycopg2.connect(database=database, user=username, password=pwd, host=address, port=portnum)
      curr = conn.cursor()
      
      # Read the character from parameter.txt
      with open("parameter.txt", "r") as file:
          char = file.readline().strip()
    
      
      # Execute the query to get the sum of scores
      query = f""" SELECT SUM(m.host_team_score) AS total_score FROM matches m
      JOIN teams t ON m.host_team_id = t.team_id WHERE m.host_team_score > m.guest_team_score
      AND t.name LIKE  '{char}%'; """
      curr.execute(query)
      result = curr.fetchone()
      for row in result:
          print(calculate(row))
      
      
  except (Exception, psycopg2.DatabaseError) as error:
      print(error)
  finally:
      if conn is not None:
          conn.close()

# Call the function
connect_db(database, user, password, host, port)
```
